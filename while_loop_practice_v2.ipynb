Q1. Print numbers from 1 to 20 using while loop.

Q7. Check if a number is palindrome.

Approach:
- Reverse number
- Compare with original

```python
#write answer here
var= int(input())
new=var
res=0
while var>0:
  temp = var % 10
  res = res*10 + temp
  var = var // 10
print(res)
if new==res:
  print("Palindrome")
else:
  print("Not Palindrome")
```
